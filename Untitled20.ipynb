**1.1 Importamos las librerias necesarias:**

```python
import pandas as pd
from scipy.stats import pearsonr
from sklearn.datasets import load_breast_cancer
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import confusion_matrix
import itertools
from sklearn.model_selection import train_test_split
from tabulate import tabulate
```

**1.2 Cargamos el conjunto de datos**

```python
data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
df
```

**1.3 Mostramos la matriz de correlación**

```python
corr_matrix = df.corr()

plt.figure(figsize=(14,12))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Matriz de Correlación Completa")
plt.tight_layout()
plt.show()
```




**1.4 Correlación de todas las variables vs target**





```python
corr_matrix = df.corr()
corr_target_pct = corr_matrix[['target']] * 100
corr_target_pct = corr_target_pct.sort_values(by='target', ascending=False)

plt.figure(figsize=(5,12))
sns.heatmap(corr_target_pct, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlación porcentual de todas las variables con Target")
plt.tight_layout()
plt.show()
```






**1.5 Tabla de Correlación con target + p-valor + Veredicto**








```python
var_names = []
corr_values = []
p_values = []
significancia = []

for col in df.columns[:-1]:
    corr, p_val = pearsonr(df['target'], df[col])
    var_names.append(col)
    corr_values.append(corr)
    p_values.append(p_val)
    significancia.append("significativa" if p_val < 0.05 else "NO significativa")

tabla_pvalores = pd.DataFrame({
    "Variable": var_names,
    "Correlación": corr_values,
    "P-valor": p_values,
    "Significancia": significancia
})


tabla_pvalores = tabla_pvalores.reindex(tabla_pvalores["Correlación"].abs().sort_values(ascending=False).index)

print("\n P-valores de Pearson con significancia:\n")
pd.set_option('display.float_format', '{:.5f}'.format)
print(tabla_pvalores)
```

Parte 2: Aplicamos el primer estudio

```python
def sigmoide(valor):
    return np.where(valor >= 0, 1 / (1 + np.exp(-valor)), np.exp(valor) / (1 + np.exp(valor)))

def sigmoide_s(z):
    z_recortado = np.clip(z, -500, 500)
    return sigmoide(z_recortado)
```

```python
def calcular_error(parametros_linea, puntos, etiquetas):
    logits = np.dot(puntos, parametros_linea)
    probabilidades = sigmoide_s(logits)
    eps = 1e-12
    probabilidades = np.clip(probabilidades, eps, 1 - eps)
    entropia_cruzada = -(np.dot(etiquetas.T, np.log(probabilidades)) + np.dot((1 - etiquetas).T, np.log(1 - probabilidades)))
    return (entropia_cruzada / puntos.shape[0]).item()
```

```python
def calcular_gradiente(parametros_linea, puntos, etiquetas):
    logits = np.dot(puntos, parametros_linea)
    probabilidades = sigmoide_s(logits)
    gradiente = np.dot(puntos.T, (probabilidades - etiquetas)) / puntos.shape[0]
    return gradiente
```

```python
def criterio_paro_norma_gradiente(gradiente, umbral_gradiente):
    return np.linalg.norm(gradiente) < umbral_gradiente
```

```python
def descenso_gradiente(puntos, etiquetas, parametros_linea,t_paso_inicial, umbral_gradiente, max_iteraciones):
    iteracion = 0
    gradiente = calcular_gradiente(parametros_linea, puntos, etiquetas)
    resultados = []
    normas_gradiente = []

    while iteracion < max_iteraciones and not criterio_paro_norma_gradiente(gradiente, umbral_gradiente):
        xk = parametros_linea.copy()
        parametros_linea = parametros_linea - t_paso_inicial * gradiente
        xk1 = parametros_linea.copy()
        gradiente = calcular_gradiente(parametros_linea, puntos, etiquetas)

        resultados.append({
            'Iteración': iteracion + 1,
            '||xk+1 - xk||': float(np.linalg.norm(xk1 - xk)),
            '||∇ f(xk)||': float(np.linalg.norm(gradiente)),
            'f(xk)': calcular_error(xk, puntos, etiquetas)
        })
        normas_gradiente.append(np.linalg.norm(gradiente))
        iteracion += 1

    return parametros_linea, resultados, normas_gradiente
    print(resultados)
```

```python
def calcular_exactitud(predicciones, etiquetas):
    return np.mean(predicciones.flatten() == etiquetas.flatten())
```

```python

def graficar_matriz_confusion(cm, nombres_objetivo, titulo='Matriz de Confusión', cmap=None, normalizar=True):

    exactitud = np.trace(cm) / float(np.sum(cm)) if np.sum(cm) > 0 else 0.0
    error = 1 - exactitud


    if cmap is None:
        cmap = plt.get_cmap('Blues')

    plt.figure(figsize=(5, 5))


    if normalizar:
        with np.errstate(all='ignore'):
            suma_filas = cm.sum(axis=1)[:, np.newaxis]
            suma_filas[suma_filas == 0] = 1
            cm_normalizada = cm.astype('float') / suma_filas
        cm_mostrar = cm_normalizada
    else:
        cm_mostrar = cm


    plt.imshow(cm_mostrar, interpolation='nearest', cmap=cmap)
    plt.title(titulo)
    plt.colorbar()


    if nombres_objetivo is not None:
        marcas = np.arange(len(nombres_objetivo))
        plt.xticks(marcas, nombres_objetivo, rotation=45)
        plt.yticks(marcas, nombres_objetivo)


    umbral = cm_mostrar.max() / 1.5 if normalizar else cm_mostrar.max() / 2


    for i, j in itertools.product(range(cm_mostrar.shape[0]), range(cm_mostrar.shape[1])):
        texto = "{:0.4f}".format(cm_mostrar[i, j]) if normalizar else "{:,}".format(int(cm[i, j]))
        plt.text(j, i, texto, horizontalalignment="center",
                 color="white" if cm_mostrar[i, j] > umbral else "black")

    plt.tight_layout()
    plt.ylabel('Etiqueta Verdadera')
    plt.xlabel('Etiqueta Predicha\nExactitud={:0.4f}; Error={:0.4f}'.format(exactitud, error))
    plt.show()
```

```python
datos = load_breast_cancer()
X, Y = datos.data, datos.target.reshape(-1, 1)


X = (X - X.mean(axis=0)) / X.std(axis=0)

X = np.hstack([X, np.ones((X.shape[0], 1))])


X_entrenamiento, X_prueba, Y_entrenamiento, Y_prueba = train_test_split( X, Y, test_size=0.5, random_state=42)
```

**Entrenamiento**

```python
t_paso_inicial = 0.001
umbral_gradiente = 1e-6
max_iteraciones = 20000


parametros_linea = np.random.randn(X_entrenamiento.shape[1], 1) * 0.01


parametros_optimizados, resultados, normas_gradiente = descenso_gradiente( X_entrenamiento, Y_entrenamiento, parametros_linea,
    t_paso_inicial, umbral_gradiente, max_iteraciones
)
```

```python
probabilidades_optimizadas = sigmoide_s(np.dot(X_prueba, parametros_optimizados))


predicciones = (probabilidades_optimizadas >= 0.5).astype(int)

exactitud = calcular_exactitud(predicciones, Y_prueba)
print(f'Exactitud: {exactitud:.4f}')
```

```python
matriz_confusion = confusion_matrix(Y_prueba.flatten(), predicciones.flatten())


graficar_matriz_confusion( matriz_confusion, nombres_objetivo=["Maligno", "Benigno"],normalizar=False)
```
